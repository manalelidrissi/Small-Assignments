# Read the file Got_1_1.txt and Got_1_3.txt  and tokenize it into sentences and words. 

```python
from nltk import word_tokenize, sent_tokenize
```

```python
sentences = []
words = []

files = ['Got_1_1.txt','Got_1_3.txt']

for file in files:
    with open('./data_files/'+file) as fs:
        for line in fs:
            sents = sent_tokenize(line.strip())
            w = word_tokenize(line.strip())
            sentences+=sents
            words+=w
```

```python
words[:10]
```

```python
sentences[:5]
```

# Clean the word tokens by removing the punctuations and stopwords. Also convert each word to lower case.

```python
import string
puncts = string.punctuation

from nltk.corpus import stopwords
s_w = stopwords.words('english')

words = [w.lower() for w in words if w not in puncts and w.lower() not in s_w]
```

```python
words[:10]
```

# Perform Stemming on the word tokens

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

stem_words = [ps.stem(w) for w in words]
```

```python
stem_words[:10]
```

# Word2Vec

Train a word2vec model using the sentences in the above text. The dimensions of the word vectors should be 200. You do not need to remove stopwords or perform stemming but remove the punctuations (obtain the top 10 results).
Check for the following semantic test examples - 

1. eddard jon cersei ?

2. arya sansa tyrion ? 


```python
sents = []
for s in sentences:
    word_list = [w.lower() for w in word_tokenize(s) if w not in puncts]
    sents.append(word_list)
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
model = Word2Vec(sents,size=50)
```

```python
model.wv.most_similar(positive=['cersei','jon'],negative=['eddard'],topn=10)
```

```python
model.wv.most_similar(positive=['tyrion','sansa'],negative=['arya'],topn=10)
```

```python
model.wv.most_similar(positive=['tyrion','sansa','arya'],topn=10)
```
