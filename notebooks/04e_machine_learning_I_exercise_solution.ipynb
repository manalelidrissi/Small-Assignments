# Exercise for Classification

```python
import pandas as pd
```

```python
df = pd.read_csv ("./data_files/credit-g.csv", na_values="?")
df
```

## Goal:
We will learn and compare predictors for the class attribute (will a person get a credit or not). We will do this step-by-step

## Data exploration:
How many features and instances does this dataset have? What is the distribution of the target class?


```python
# Number of instances
len (df)
```

```python
#Number of attributes:
len (df.columns)
```

```python
df["class"].value_counts(normalize=True)
```

## Create a numpy array for the target class
Remember that the array should contain only 1 and 0

```python
from sklearn import preprocessing
y = preprocessing.LabelEncoder().fit_transform(df["class"].values)
y
```

## Create the numpy arrays of feature values
Split categorical variables into binary ones!


```python
df_features = df.drop(["class"], axis=1)
df_features = pd.get_dummies(df_features)
df_features
```

```python
    X = df_features.values
```

## Split the data randomly into 50% training and 50% test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

## Learn a Random Forest classifier on the training data

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
clf.fit (X_train, y_train)
```

```python
sum(clf.predict(X_train) == y_train) / len(y_train)
```

## Predict the class labels for the test dataset
Predict either absolute (1 or 0) or as probabilities

```python
clf.predict (X_test)
```

```python
clf.predict_proba (X_test)
```

## Which features to influence the  classifier most?

```python
fi = clf.feature_importances_
feature_importances_df = pd.DataFrame ({"Feature" : df_features.columns, "importance": fi})
feature_importances_df.sort_values("importance", ascending=False).head(5)
```

## What is the accuracy of this classifier on the _Training Data_?

```python
(clf.predict(X_train) == y_train).sum() / len(y_train)
```

## What is the accuracy of this classifier on the _Test Data_?

```python
(clf.predict(X_test) == y_test).sum() / len(y_test)
```

## What is the mean 10-fold cross-validated accuracy of this classifier?

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf, X, y, cv=10, scoring="accuracy").mean()
```

## How long did the cross-validation take?

```python
%time cross_val_score(clf, X, y, cv=10, scoring="accuracy")
```

## Compare this with a variety of other classifiers of your choice

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
clf_names = ["k-Nearest Neighbors", 
             "Decision Tree", 
             "Random Forest", 
             "Neural Net", 
             "Naive Bayes",
             "AdaBoost"]
clfs = [KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    MLPClassifier(),
    GaussianNB(),
    AdaBoostClassifier()]
```

```python
for name, clf in zip(clf_names, clfs):
    acc = cross_val_score(clf, X, y, cv=10, scoring="accuracy").mean()
    print ("Accuracy of %s according to cross-validation: %s" % (name, acc))
```

# Congressional Data:
This data set includes votes for each of the U.S. House of Representatives Congressmen on the 16 key votes identified by the CQA. The CQA lists nine different types of votes: voted for, paired for, and announced for (these three simplified to yea), voted against, paired against, and announced against (these three simplified to nay), voted present, voted present to avoid conflict of interest, and did not vote or otherwise make a position known (these three simplified to an unknown disposition).

Source: https://archive.ics.uci.edu/ml/datasets/congressional+voting+records

Task: Build a classifier that can predict the party of a congress member based on his decision. use the last 50 instances of the dataset as a test set

```python
df = pd.read_csv ("./data_files/vote.csv", na_values="?")
df
```

```python
y = preprocessing.LabelEncoder().fit_transform(df["party"].values)
df_features = df.drop(["party"], axis=1)
df_features = pd.get_dummies(df_features)
X = df_features.values
```

```python
y_train = y[:-50]
y_test = y[-50:]
X_train = X[:-50]
X_test = X[-50:]
```

```python
from sklearn.metrics import accuracy_score
```

```python
for name, clf in zip(clf_names, clfs):
    clf.fit (X_train, y_train)
    acc = accuracy_score(y_test, clf.predict(X_test))
    print ("Accuracy of %s according to cross-validation: %s" % (name, acc))
```
