## Home Assignment like question:
**The goal of this task is, that you get familiarized with the structure of task you are provided in the home assignments!**
It is not necessarly of the same difficulty.

Write a function select_values that takes in a dataframe as first argument and a dict as second arguments.
The keys of the dict correspond with columns, while the values of the dict correspond to values of that column.
It returns the dataframe such that only those rows are selected where the column has the specified value

```python
def select_values(df_in, d):
    df=df_in
    for key,value in d.items():
        df = df[df[key]==value]
    return df
```

```python
# create toy dataset
df_test = pd.DataFrame.from_dict({'A':[1,2,3,3,2,1], 'B' : ['a','b','a','b','a','b']})
# Apply some tests
assert(all(select_values(df_test, {'A' : 3}).index == [2, 3]))
assert(all(select_values(df_test, {'B' : 'a'}).index == [0, 2, 4]))
assert(all(select_values(df_test, {'A' : 3, 'B' : 'a'}).index == [2]))
```
